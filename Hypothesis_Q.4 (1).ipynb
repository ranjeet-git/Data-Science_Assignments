```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
df=pd.read_csv(r"C:\Users\k19ra\Downloads\Costomer+OrderForm.csv")
df
```

```python
df_melt = pd.melt(df.reset_index(), id_vars=['index'], value_vars=['Phillippines', 'Indonesia', 'Malta', 'Malta'])
# replace column names
df_melt.columns = ['index', 'Country', 'value']
```

```python
df_melt
```

```python
import researchpy as rp
```

```python
table,results = rp.crosstab(df_melt['Country'],df_melt['value'],test='chi-square')
```

```python
results
```

```python
alpha = 0.05
pvalue =0.2771
```

```python
if pvalue < alpha:
    print("Ho is rejected and H1 is accepted")
else:
    print("H1 is rejected and H0 is accepted")
```
